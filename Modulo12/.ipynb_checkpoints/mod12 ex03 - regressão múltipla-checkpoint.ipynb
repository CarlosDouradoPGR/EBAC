````python
import pandas as pd
import numpy as np
from seaborn import load_dataset

import matplotlib.pyplot as plt
import seaborn as sns
import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf
````

### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
tips = load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
tips.shape
````

#### 1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  

Não encontrei a variável diner no DataFrame

````python
y, X = patsy.dmatrices("tip ~ sex + smoker + time +  net_bill", tips[:100])
````

````python
X
````

````python
X[:, 4]
````

````python
modelo = sm.OLS(y,X).fit()
````

````python
modelo.summary()
````

````python
sns.scatterplot(x= X[:, 4], y=modelo.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

#### 2. Remova as variáveis não significantes.

````python
y, X = patsy.dmatrices("tip ~ time + net_bill", data=tips[:100])
````

````python
X
````

````python
reg = sm.OLS(y, X).fit()
````

````python
reg.summary()
````

### 3. observe o gráfico de resíduos em função de ```net_bill```  

````python
sns.scatterplot(x= X[:, 2], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

### 4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
reg = smf.ols("tip ~ np.log(net_bill) + time", data= tips[:100]).fit()
reg.summary()
````

````python
# Gráfico do polinomio do segundo grau
sns.scatterplot(x= tips.loc[:100, "net_bill"], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
reg = smf.ols("tip ~ np.log(net_bill) + time", data= tips[:100]).fit()
reg.summary()
````

````python
# Gráfico com o logaritimo de net_bill

sns.scatterplot(x= tips.loc[:100, "net_bill"], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

#### Para ambos os casos, o r-squared diminui, mas se analisar somente os gráficos de residos o que apresenta um melhor gráfico é o do polinomio do segundo grau, mas a melhor regressão continua sendo a primeira, sem nem uma das mudanças

````python
# regressão 1

reg_1 = smf.ols("tip ~ net_bill + time", data= tips[:100]).fit()
reg_1.summary()
````

````python
# Gráfico de residos com a melhor regresão encontrada até esse ponto

sns.scatterplot(x= tips.loc[:100, "net_bill"], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

## Limpando os dados

````python
df = pd.read_csv("previsao_de_renda.csv")
````

````python
df.head()
````

````python
df.info()
````

````python
#Removendo colunas que não serão úteis
df.drop(columns="Unnamed: 0", inplace=True)
````

````python
# Transformando coluna data_ref de object para datetime
df["data_ref"] = pd.to_datetime(df["data_ref"])
````

````python
# Como a variável explicativa será tempo_emprego, iremos retirar os valores missing
df.dropna(inplace=True)
````

````python
df.info()
````

````python
df.head()
````

## Montando o modelo

````python
y, X = patsy.dmatrices("renda ~ tempo_emprego", df)
reg = sm.OLS(y, X).fit()
reg.summary()
````

````python
sns.scatterplot(x=X[:, 1], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

## Regressão múltipla

````python
y, X = patsy.dmatrices("""np.log(renda) ~ tempo_emprego + 
                       idade +
                       qtd_filhos + 
                       sexo + 
                       tipo_renda +
                       tipo_residencia""", df)
````

````python
#X[:, [0, 1, 3, 5,11,12,13]]
````

````python
reg = sm.OLS(y, X).fit()
reg.summary()
````

````python
sns.scatterplot(x=X[:, 11], y=reg.resid)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
fig, ax = plt.subplots(figsize=(10,6))

sns.scatterplot(ax=ax, x=X[:, 11], y=y[:, 0], color= "b")
sns.scatterplot(ax=ax, x=X[:, 11], y=reg.predict(), color="r")
ax.set_ylabel('Renda')
ax.set_xlabel("Tempo de emprego")
ax.set_title("Predict x True_train")

plt.show()
````

````python
sns.boxplot(x=df["sexo"], y=reg.resid)
````

````python
sns.boxplot(x=df["tipo_renda"], y=reg.resid)
````

- se utilizarmos apenas o r-squared como paramentro para a avaliação do modelo, esse modelo seria o ideal
