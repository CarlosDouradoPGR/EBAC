# Regressão 01 - tarefa 02 - transformações em X e Y

````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf
````

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
tips = load_dataset("tips")
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips["net_bill"] = tips["total_bill"] - tips["tip"]
tips.head()
````

````python
data_train = tips[:80]
````

````python
#Criando a regressão
reg = smf.ols('tip_pct ~ net_bill', data = data_train).fit()
reg.summary()
````

````python
# Plotando a relação entre os valores fitados e "net_bill"

sns.regplot(x=data_train["net_bill"], y=reg.fittedvalues, line_kws={'alpha': 0.5})
sns.lineplot(x=[i for i in range(0, 50, 5)], y=[np.mean(reg.fittedvalues)] * 10, linestyle='--', color="g")
plt.text(x=36, y=0.2, s="linha de média", color="g")
plt.plot()
````

````python
# Visualisando os residos
data_train["resid"] = reg.resid

sns.scatterplot(x="net_bill", y="resid", data=data_train)
plt.yticks(np.arange(-1, 1.1, 0.2).tolist())
plt.axhline(y=0, color='r', linestyle='--')
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
# Ajustando a regressão v1
reg = smf.ols('tip_pct ~ np.log(net_bill+0.001)', data = data_train).fit()
reg.summary()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.
