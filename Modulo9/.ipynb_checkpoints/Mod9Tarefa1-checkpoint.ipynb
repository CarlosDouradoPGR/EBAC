# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc.groupby(["munResNome"])[["IDADEPAI","IDADEPAI"]].mean().rename_axis("Municipio").reset_index()
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

R: Eu nasci no dia 29/08/2000

```python
sinasc["DTNASC"] = pd.to_datetime(sinasc["DTNASC"])
```

```python
sinasc_dt = sinasc[sinasc["DTNASC"] == pd.to_datetime("2019-08-29") ]
```

```python
sinasc_dt.groupby(["ESCMAE", "SEXO"])["PESO"].mean().to_frame()
```

```python
sinasc_dt.head()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc.groupby("munResNome")["munResNome"].count().sort_values().head(3)
```

Acredito que ``Município ignorado - RO`` tenha sido um erro de digitação, podemos considerar com um outline, então o municipio que nasceu menos bebes no ano de 2019 foi castanheiras

```python
sinasc_idde = sinasc[sinasc["munResNome"] == "Castanheiras"]
```

```python
sinasc_idde.groupby("munResNome")[["IDADEPAI", "IDADEMAE"]].agg([np.mean, min, max])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc[sinasc["DTNASC"].dt.month == 3].groupby("munResNome")["munResNome"].count().sort_values(ascending=False).head(5)
```

```python
sinasc[sinasc["munResNome"] == "Porto Velho"].groupby("munResNome")[["QTDFILVIVO", "IDADEPAI"]].agg([np.mean, min, max])
```

### 1º insight

O gráfico abaixo nos da uma informação sobre qual tempo de getação e o tipo de parto dos bebes mais saúdaveis de acordo com o apgar

```python
sinasc.groupby(["PARTO", "GESTACAO"])[["APGAR1", "APGAR5"]].mean().sort_values(by="APGAR1").plot(figsize=(18,6))
```

### 2º insight

```python
sinasc_pai = sinasc.dropna(subset=["IDADEPAI"])
```

```python
sinasc_pai.groupby(["ESCMAE"])[["IDADEMAE", "IDADEPAI"]].agg([np.mean, max, min]).sort_values(by=("IDADEMAE", "mean"))
```

No DataFrame acima podemos ter uma informação errada, se olharmos apenas para idade, podemos dizer que uma mãe com 12 anos ou mais de escolaridade uma média de diade próxima a que não tem nem uma escolaridade.

```python
sinasc_pai.groupby(["ESCMAE"])[["IDADEMAE", "IDADEPAI","QTDFILVIVO" ]].agg([np.mean, max, min]).sort_values(by=("IDADEMAE", "mean"))
```

Mas quando levamos em conta a Quantidade de Filho Vivo(QTDFILVIVO) percebemos que enquanto as mães com 12 anos ou mais de escolaridade estão no primeiro filho as mãe sem nenhuma escolaridade já tem a média de 5 filhos vivos.
