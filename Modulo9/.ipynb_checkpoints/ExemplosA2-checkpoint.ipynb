```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme()  
%matplotlib inline
```

```python
df = pd.read_csv("SINASC_RO_2019.csv")
```

```python
df.info()
```

```python
sinasc_1 = df[['DTNASC', "SEXO", "PESO", "ESCMAE", "GESTACAO", "IDADEMAE"]]
```

```python
sinasc_1.info()
```

#### A) Quantidade de bebês ao longo do ano

```python
sinasc_1dt = sinasc_1.groupby('DTNASC')["IDADEMAE"].count().plot(figsize=(18,6))
plt.ylabel("Quantidade de nascimento")
```

#### B) Quantidade de bebês fem e masc ao longo do ano

```python
sinasc_1["SEXO"].value_counts()
```

```python
fem = sinasc_1[sinasc_1["SEXO"] == 'Feminino'].groupby("DTNASC")["IDADEMAE"].count()
masc = sinasc_1[sinasc_1["SEXO"] == "Masculino"].groupby("DTNASC")["IDADEMAE"].count()
```

```python
plt.close("all")
fig, axes = plt.subplots(figsize=(18,6))

fem.plot()
masc.plot()
plt.ylabel("Quantidade de nascimento");
```
