#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

Novamente, assim como em ``ESTCIVMAE``, o 9 é cosiderado ignorado em ``CONSULTAS``

Número de consultas de pré-natal:
- 1: Nenhuma
- 2: de 1 a 3
- 3: de 4 a 6
- 4: 7 e mais
- 9: Ignorado

```python
consultas = ["Nenhuma", "1 a 3", "4 a 6", "mais de 7"]
```

```python
df_3 = df[["APGAR5", "CONSULTAS"]]
```

```python
df_3 = df_3[df_3["CONSULTAS"] != 9]
```

```python
df_3["CONSULTAS"].value_counts()
```

```python
media = list()

for i in range(1, 5):
    media.append(df_3[df_3["CONSULTAS"] == i]["APGAR5"].mean())
```

```python
fig, ax = plt.subplots()

a = sns.barplot(x=consultas, y=media, palette="viridis")
ax.set_xlabel("Numero de Consultas no Pré-Natal")
ax.set_ylabel("Média do APGAR5")
for i in range(len(media)):
    ax.text(i, media[i], f'{media[i]:.2f}', ha='center', va='bottom')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

Relembrando que retirei dos gráficos os valores ignorados(``9``)

```python
df_4 = df[['SEXO', 'GESTACAO', 'ESTCIVMAE','CONSULTAS']]
```

```python
df_4 = df_4[(df_4['ESTCIVMAE'] != 9) & (df_4['CONSULTAS'] != 9) & (df_4['GESTACAO'] != 9)]
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,10))

sns.countplot(ax = axes[0,0], x="SEXO", data=df_4, palette={'Feminino':'pink', 'Masculino':'blue'})
sns.countplot(ax = axes[0,1], x="GESTACAO", data=df_4)
sns.countplot(ax = axes[1,0], x="ESTCIVMAE", data=df_4)
sns.countplot(ax = axes[1,1], x="CONSULTAS", data=df_4)
axes[0,1].tick_params(axis="x", rotation = 65)
for i in [axes[0,0],axes[0,1], axes[1,0], axes[1,1]]:
     i.set_ylabel(" Freqência")
    


#ajustar o espaçamento
plt.subplots_adjust(wspace=0.4, hspace=0.8)
```
