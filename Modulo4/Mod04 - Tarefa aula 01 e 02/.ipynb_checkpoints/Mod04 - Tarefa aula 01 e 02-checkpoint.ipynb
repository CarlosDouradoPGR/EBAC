#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

# Seu código aqui

ano = np.array([1900, 1920, 1940, 1960, 1980, 2000, 2020])
anomalia_termica = np.array([-0.08, -0.27, 0.12, -0.03,  0.26, 0.40, 1.02])


serie = pd.Series(anomalia_termica)
serie
```

#### 2) Coloque os anos nos índices conforme a tabela.

```python

# seu código aqui
serie = pd.Series(anomalia_termica, index=ano)
serie
```

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui

media = list()

for i in range(len(df)):
    media.append(np.sum(df.iloc[i]) / 3)
    
df["media"] = media
df
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui
log_e = list()

for i in range(len(df)):
    log_e.append(np.log(df.iloc[i, 3]))
    
    
```
