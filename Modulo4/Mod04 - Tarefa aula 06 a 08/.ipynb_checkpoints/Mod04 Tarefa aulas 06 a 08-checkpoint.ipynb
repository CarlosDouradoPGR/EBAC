Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?

#### 1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), e que devolve os dados da consulta no formato DataFrame.

````python
url = "https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl"

rqst.get(url)
````

````python
s = rqst.get(url)

data_js = s.json()
````

````python
df = pd.DataFrame(data_js["items"])
````

````python
df.loc[(df["tipo_interessado"] == "Estado") & (df["uf"] == "PA")].head()
````

````python
# Função
def Extrair_dados(df=pd.DataFrame, UF=str, tipo_interessado=str):
   return df.loc[(df["uf"] == UF) & (df["tipo_interessado"] == tipo_interessado)]

````

````python
Extrair_dados(df=df, UF="PA", tipo_interessado="Município").head()
````

#### 2.Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?

````python
# 2) Seu código aqui
df_mg = Extrair_dados(df, "MG", "Estado")
````

````python
df_mg["status"].value_counts()
````

Não existe o status *Arquivado por decurso de prazo* no dataframe solicitado

#### 3.Qual é o município da Bahia com mais solicitações deferidas?

````python
#3 Seu código aqui
df_ba = Extrair_dados(df,"BA", "Município" )
df_ba.head()
````

````python
df_ba["status"].value_counts()["Deferido"]
````

#### 4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# 4) Seu código aqui
ba_estado = Extrair_dados(df, "BA", "Estado")
````

````python
ba_estado.head()
````
